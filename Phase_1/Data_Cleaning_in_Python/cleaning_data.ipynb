```python
numeric_values=data.select_dtypes(include=[float, int]).columns
```

```python
fill_values=data[numeric_values].fillna(data[numeric_values].median())
fill_values.info()
```

```python
fill_values=data
fill_values.info()
```

```python
category_values=data.select_dtypes(include=['object']).columns
category_values
```

```python
fill_object=data[category_values].fillna(data[category_values].mode().iloc[0])
fill_object.info()
```

```python
cleaned_data=pd.concat([fill_values , fill_object], axis=1)
cleaned_data.info()
```

```python
import csv
cleaned_data.to_csv("Electric_Vehicle_Population_Data.csv")
```
